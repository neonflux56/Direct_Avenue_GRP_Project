# How does estimated reach and audience size vary by Station Category?

```python
grouped_preds = dat.groupby('Station_Category').mean()[['Estimated_Audience_Size', 'Estimated_REACH']].reset_index()
vals = dict(grouped_preds.loc[:, ['Station_Category', 'Estimated_Audience_Size']].values.tolist())
ticks = dict(sorted((value, key) for (key, value) in vals.items())).values()

bp = sns.barplot(x='Station_Category', y='Estimated_Audience_Size', order=list(ticks), palette=("Blues_d"), 
                 data=dat)
plt.title('Estimated Audience Size by Station Category')
plt.xlabel('Station Category')
plt.ylabel('Estimated Audience_Size')
plt.show()
```

```python
sns.barplot('Station_Category', 'Estimated_REACH', order=list(ticks), palette=("Blues_d"), 
                 data=dat)
plt.title('Estimated Reach Size by Station Category')
plt.xlabel('Station Category')
plt.ylabel('Estimated Reach')
plt.show()
```

# What if we plot a relationship between spot cost and estimated reach?

```python
filt = dat[(dat['Estimated_REACH']<2*dat['Estimated_REACH'].mean()) & (dat['Spot_Cost']<2*dat['Spot_Cost'].mean())]
fin = filt.sample(frac=0.05)
sns.lmplot('Spot_Cost', 'Estimated_REACH', data=fin, hue='Market', height=8, aspect=1.3)
plt.title('Spot Cost/Reach Relationship')
plt.xlabel('Spot Cost')
plt.ylabel('Estimated Reach')
plt.show()
```

# Time Series Plot by Month

```python
import datetime

dat['Month_num'] = [datetime.datetime.strptime(x, '%B').month for x in dat['Month'].values]
lp = sns.lineplot(x='Month_num', y='Estimated_REACH', data=dat, err_style='bars', hue='Market')
lp.set_xticks(list(range(1,13)))
lp.set_xticklabels(list(range(1, 13)))
plt.title('Estimated Reach by Month')
plt.xlabel('Month')
plt.ylabel('Reach')
plt.show()
```

# Estimated Reach for Access Code and Spot Length

```python
fig, axes = plt.subplots(1, 2, sharey=True)

sns.barplot('Access', 'Estimated_REACH', data=dat[dat['Access']!='15'], ax=axes[0])
sns.barplot('Length', 'Estimated_REACH', data=dat, ax=axes[1])
```
